```python
def grade_category(num):
    if num < 8:
        return 'D'
    elif num in range(8, 11):
        return 'C'
    elif num in range(11, 14):
        return 'B'
    elif num >= 14:
        return 'A'

data_df['G3'] = data_df['G3'].apply(grade_category)
```

# Train a model
train a model manually

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
import torch.optim as optim
```

## Model

### Regression Model

```python
class MLP_reg(nn.Module):

    def __init__(self, feature_num, dropout=0.5):
        super(MLP_reg, self).__init__()
        self.layer = nn.Sequential(
            nn.Linear(feature_num, 60),
            nn.LeakyReLU(0.02),
            nn.Dropout(dropout),

            nn.Linear(60, 30),
            nn.LeakyReLU(0.02),
            nn.Dropout(dropout),

            nn.Linear(30, 1)
        )


    def forward(self, x):
        return self.layer(x)


model = MLP_reg(len(dataset.dummy_features))
```

### Classification Model

```python
class MLP_class(nn.Module):
    
    def __init__(self, feature_num, class_num, dropout=0.5):
        super(MLP_class, self).__init__()
        self.layer = nn.Sequential(
            nn.Linear(feature_num, 60),
            nn.LeakyReLU(0.02),
            nn.Dropout(dropout),

            nn.Linear(60, 30),
            nn.LeakyReLU(0.02),
            nn.Dropout(dropout),

            nn.Linear(30, class_num),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.layer(x)

model = MLP_class(len(dataset.dummy_features), len(dataset.dummy_target))
```

## Prepare Dataset

```python
train_X = torch.from_numpy(
    dataset.get_train_X().values).float()
train_y = torch.from_numpy(
    dataset.get_train_y().values).float()
test_X = torch.from_numpy(
    dataset.get_test_X().values).float()
test_y = torch.from_numpy(
    dataset.get_test_y().values).float()

train_dataset = TensorDataset(train_X, train_y)
test_dataset = TensorDataset(test_X, test_y)
```

## Training

```python
epoch = 100
lr = 0.001
batch_size = 16

optimiser = optim.Adam(model.parameters(), lr=lr)
criterion = nn.BCELoss()
```

```python
def evaluate():
    model.eval()
    data_loader = DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=True)

    pred_class = np.array([])
    target_class = np.array([])

    for x, target in data_loader:
        pred = model(x)
        pred = pred.argmax(axis=1).numpy()
        target = target.argmax(axis=1).numpy()
        target_class = np.concatenate((target_class, target))
        pred_class = np.concatenate((pred_class, pred))

    return (target_class == pred_class).mean()
```

```python
def train():
    data_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
    
    for e in range(epoch):
        model.train()

        for x, target in data_loader:
            optimiser.zero_grad()

            pred = model(x)
            err = criterion(pred, target)
            err.backward()
            optimiser.step()
        
        print("Epoch: {}, loss={:.3f}, test accuracy={:.3f}".format(e, err, evaluate()))
```

```python
train()
```
